# Monopoly

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, fixed
%matplotlib inline
```

```python
A = 1000
b = 1
F = 0
c = 200
d = 0.02

def P(Q):
    return A - b*Q

def TR(Q):
    return P(Q)*Q

def MR(Q):
    return A - 2*b*Q

def TC(Q):
    return c*Q + d*Q**2

def MC(Q):
    return c + 2*d*Q

def PR1(q1,q2):
    return P(q1+q2)*q1 - TC(q1)
```

```python
def monop(q, pr=False):
    pq = P(q)
    f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8,8))
    ax1.set_ylim(0,A*1.1)
    ax1.set_xlim(0,len(Q))
    ax1.plot(Q, P(Q))
    ax1.scatter(q, pq)
    ax1.plot(Q,MR(Q))
    ax1.scatter(q,MR(q))
    ax1.hlines(pq,xmin=0, xmax=q)
    ax1.vlines(q,ymin=0, ymax=pq)
    ax1.fill_between(q,pq*np.ones(int(q)),np.ones(int(q))*0,facecolor='gold'   )
    ax2.plot(Q,TR(Q))
    ax2.scatter(q,TR(q))
    ax2.set_ylim(0,260000)

    if pr:
        ax2.plot(Q,TR(Q)-TC(Q))
        ax2.scatter(q,TR(q)-TC(q))
        ax1.plot(Q,MC(Q))
        ax2.plot(Q,TC(Q))
    
   #plt.show();
```

```python
interact(monop, q=(0,A,10))
```
