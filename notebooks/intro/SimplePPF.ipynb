```python
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact
```

The following is just styling for the graph

```python
plt.style.use('bmh')
plt.rcParams["figure.figsize"] = [7,7]
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False
plt.rcParams["font.size"] = 18
```

```python
LE = 30    
MPL_F = 1.0
MPL_B = 1.0
```

```python
def Q(mpl):
    plt.plot([0,30],[0,30*mpl])
    plt.title("Production of Fries", fontsize = 18)
    plt.xlim(0, 40) 
    plt.ylim(0, 150)
    plt.xlabel("Labor input LF")
    plt.ylabel("Output QF -- Fries")
```

```python
def ppf(mplF = 1, mplB=1, LE = LE, wp =1, price=False):
    QFmax = mplF * LE
    QBmax = mplB * LE
    print('Domestic relative price:')
    print(' {:5.2f} Burgers per Fries'.format(mplB/mplF))
    plt.plot([0, QFmax], [ QBmax,0 ], marker='o',clip_on=False)
    plt.plot([0,LE*3],[0,LE*3], linestyle =":")
    
    if price:
        print("At a world price of {} coconut per fish this country:".format(wp))
        if (wp < mplB/mplF):
            plt.plot([0, QBmax*wp], [QBmax,0 ], 'r--',clip_on=False)
            print('   specialize at BURGERS')
        elif (wp > mplB/mplF):
            plt.plot([0, QFmax], [QFmax*wp,0 ], 'r--',clip_on=False)
            print('   sepcialze at FRIES')
        else:
            print('   has NO TRADE')
        
    plt.title("PPF", fontsize = 18)
    plt.xlim(0, LE*3) 
    plt.ylim(0, LE*3)
    plt.xlabel("QF -- FRIES")
    plt.ylabel("QB -- BURGERS")
    plt.show()
```

### Linear Production

The production of Fries:
$$ Q_{F} = MPl_F * L_F$$

```python
Q(mpl=2)
```

```python
interact(Q, mpl=(0,4,0.1))
```

### The PPF Model

You have 30 minutes of work each morning.  All of his attention is focused on preparing Fries and Burgers.  The linear production technologies and resource constraints are:

$$ Q_{F} = MPl_F * L_F$$

$$ Q_{B} = MPL_B * L_B$$

$$ L_{B} + L_B  = 30$$


Let's define a function for plotting Tom's Production possibility frontier.  The same function will have the option of also drawing a world price line.

Let's plot Tom's production possibility frontier at the default values:

```python
ppf(mplF = 1, mplB=2, LE = 30)
```

Let's use ipython widgets to see how the plot changes when we vary the parameters via sliders and buttons:

```python
interact(ppf, mplF=(0.5, 3, 0.5), mplB=(0.5, 3, 0.5), LE=(1,2*LE,LE/10),wp=(0.1,3,0.1));
```
