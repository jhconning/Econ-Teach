# Fishing example

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

There are two fishermen who fish in a lake.  For each hour that Bob fishes, Alfredo catches 1/4 fish per hour less (and vice-versa). If Bob fishes $e_B$ hours, and Alfredo fishes $e_A$ hours, Alfredoâ€™s catch is:

$$y_A =\left ( 15 -\frac{1}{4} e_B \right ) \cdot e_A$$

And the utility he receives from the catch (after factoring in the disutility of effort) is:

$$u_A =\left ( 15 -\frac{1}{4} e_B \right ) \cdot e_A - \frac{1}{2} e_A^2$$

```python
def yA(ea, eb):
    return (15 - (1/4)*eb)*ea

def yB(ea, eb):
    return (15 - (1/4)*ea)*eb

def uA(ea,eb):
    return yA(ea, eb) - (1/2)*ea**2

def uB(ea,eb):
    return yB(ea, eb) - (1/2)*eb**2
```

### Prisoner's dilemma

Cooperate:  10 hours

Defect: 12 hours

```python
yA(10,10), yA(12,10), yA(10,12), yA(12,12)
```

```python
(uA(10,10), uB(10,10)),(uA(10,12), uB(10,12))
```

```python
(uA(12,10), uB(12,10)),(uA(12,12), uB(12,12))
```

### Best response functions

$\frac{\delta u_A}{\delta e_A} = 15 -\frac{1}{4}e_B- e_A =0$

So the best response is:

$R_A(e_B) = 15 -\frac{1}{4}e_B$

```python
ea = np.arange(1,20)
eb = np.arange(1,20)
plt.title(r"A\'s best response to different $e_B$")
plt.xlabel(r'$e_A$')
plt.plot(ea,uA(ea, eb=0))
plt.plot(ea,uA(ea, eb=5))
plt.plot(ea,uA(ea, eb=10));
```

$$\frac{\partial u_A}{\partial e_A} = 15 - \frac{1}{4}e_B - e_A=0$$

$$R_A(e_B) = 15- \frac{1}{4}e_B$$

```python
def RA(eb):
    return 15 - (1/4)*eb

def RB(ea):
    return 15 - (1/4)*ea
```

```python
plt.xlim(0,20)
plt.ylim(0,20)
plt.plot(RA(eb),eb)
plt.plot(ea,RB(ea));
```

Symmetric Nash at: 

$$e^* = 15 - \frac{1}{4}e^*$$

$$\frac{5}{4}e^* = 15$$

$$e^* = \frac{4}{5} \cdot 15 =12$$

### Efficient

$$\max u_A(e_A, e_B)$$

subject to 

$$u_B(e_A, e_B)=\bar u$$


$$u_A =\left ( 15 -\frac{1}{4} e_B \right ) \cdot e_A - \frac{1}{2} e_A^2$$
