##  Model

#### Demand
The monopolist faces an (inverse) **Demand Function** $P(Q)$ given by:

$$
P(Q) = A - b*Q
$$

This implies that **Total Revenue** $TR(Q)$ can be written:

$$
TR(Q) = P(Q) \cdot Q = A \cdot Q - b \cdot Q^2
$$

and **Marginal Revenue** $MR(Q)=\frac{dTR(Q)}{dQ}$ is:

$$
MR(Q) = A - 2 b \cdot Q
$$

#### Production 

The monopolist has a **Total Cost curve** $TC(Q)$  given by:

$$
TC(Q) = F + c \cdot Q + d \cdot Q^2
$$

This implies that **Marginal Cost** can be written:

$$
MC(Q) = c+ 2d \cdot Q
$$

#### Monopolist's output choice

The monopolist chooses $Q$ to  maximize total profits:

$$
\Pi(Q) = TR(Q) -TC(Q)
$$

The first-order necessary condition for an interior maximum is:
    
$$
MR(Q^*) = MC(Q^*)
$$

For these functional forms, this implies:

$$
A - 2b\cdot Q^* = c+2d*Q^*
$$

```python
# Global parameters
A = 1000     # intercept of demand curve          
b = 1        # slope of demand curve
F = 0        # fixed cost
c = 200      # marginal cost c + d*Q  
d = 0.02     

def P(Q):
    return A - b*Q

def TR(Q):
    return P(Q)*Q

def MR(Q):
    return A - 2*b*Q

def TC(Q):
    return F + c*Q + d*Q**2

def MC(Q):
    return c + 2*d*Q
```

### Plots
#### Demand, Marginal Revenue and Marginal Cost curves:

```python
Q = np.arange(0,1100,1)
plt.xlim(0,1100)
plt.ylim(0,1100)

plt.plot(Q, P(Q))
plt.plot(Q,MR(Q))
plt.plot(Q, MC(Q));
```

###  Total Revenue and Total Cost Curves

```python
plt.plot(Q, TR(Q))
plt.plot(Q, TC(Q))
plt.ylim(0,270000)
plt.grid()
```

```python
def monop(q, pr=False):
    '''Plot TR, TC, MC, MR curves for a monopoly firm with demand curve P(Q) and cost function TC(Q)
       Allow user to specify arbitrary quantity q.
       The pr=True option plots the total profit curve.'''
    pq = P(q)
    f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8,8))
    ax1.set_ylim(0,A*1.1)
    ax1.set_xlim(0,len(Q))
    ax1.plot(Q, P(Q))
    ax1.scatter(q, pq)
    ax1.plot(Q, MR(Q))
    ax1.scatter(q, MR(q))
    ax1.hlines(pq,xmin=0, xmax=q)
    ax1.vlines(q,ymin=0, ymax=pq)
    ax1.fill_between(q, pq*np.ones(int(q)),np.ones(int(q))*0,facecolor='gold'   )
    ax1.set_title(r'P(Q), MR(Q), MC(Q)')
    ax2.plot(Q,TR(Q))
    ax2.scatter(q,TR(q))
    ax2.set_ylim(0,260000)
    ax2.set_title(r'TR(Q), TC(Q)')


    if pr:
        ax2.set_title(r'$TR(Q), TC(Q), \Pi(Q)$')
        ax2.plot(Q,TR(Q)-TC(Q))
        ax2.scatter(q,TR(q)-TC(q))
        ax1.plot(Q,MC(Q))
        ax2.plot(Q,TC(Q))
    
   #plt.show();
```

Move the slider to set a level of output $q$.  The $pr$ flag plots the total profit curve in the bottom plot.  Use this to manually discover the profit maximizing level.

```python
interact(monop, q=(0,A,10))
```
