# Simple Tariff

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Q = np.linspace(0,100,100)
S = Q*0.5
D = 50 - 0.5*Q
```

```python
plt.style.use('seaborn-bright')
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111) 
plt.plot(Q,S)
plt.plot(Q,D)
plt.axhline(10)
plt.axhline(15)
plt.xlim(0,30)
plt.ylim(0,30)
plt.xlabel('Q - jets')
major_ticks = np.arange(0, 101, 10)   
ax.set_xticks(major_ticks) 

plt.text(87,7,"D",fontsize=22)
plt.text(57,27,"S",fontsize=22)
plt.text(-20,10,r"$P_W$",fontsize=22)
plt.text(-30,15,r"$P_W(1+t)$",fontsize=22)
plt.rc('font', size=25)   
plt.grid()
```
